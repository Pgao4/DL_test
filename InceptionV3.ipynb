This is a very simple example of using InceptionV3 in TensorFlow and Keras to classify a image.

Load TensorFlow, Keras, InceptionV3 and Python libraries.

```python
# %tensorflow_version 2.x
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.inception_v3 import InceptionV3
from keras.applications.inception_v3 import preprocess_input
from keras.applications.inception_v3 import decode_predictions
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
print(keras.__version__)
```

Set the model to InceptionV3 and print model summary.

```python
model = InceptionV3(weights='imagenet')
```

```python
print(model.summary())
```

Load an image from the net.

```python
image=tf.keras.utils.get_file(origin='https://images.unsplash.com/photo-1532974297617-c0f05fe48bff?ixlib=rb-1.2.1&ixid=eyJhcHBfaWQiOjEyMDd9&auto=format&fit=crop&w=700&q=80',fname='car.jpg')
```

Resize image.

```python
image = load_img(image, target_size=(299, 299))
```

Show image

```python
plt.imshow(image)
```

Convert image to Numpy array

```python
image = tf.keras.preprocessing.image.img_to_array(image)
```

Reshape data for the model.

```python
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
```

Preprocess data for the InceptionV3 model.

```python
image = preprocess_input(image)
```

Predict the probability across all output classes.

```python
p = model.predict(image)
```

Convert the probabilities to class labels.

```python
label = decode_predictions(p)
```

Print the label.

```python
print(label)
```
